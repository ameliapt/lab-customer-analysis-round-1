```python
import pandas as pd
import numpy as np
```

Read the 3 files into python as dataframes

```python
file1_df = pd.read_excel('files_for_lab/excel_files/file1.xlsx')
file2_df = pd.read_excel('files_for_lab/excel_files/file2.xlsx')
file3_df = pd.read_excel('files_for_lab/excel_files/file3.xlsx')

display(file1_df)
display(file2_df)
display(file3_df)
```

Show the DataFrame's shape

```python
print('File1 shape:', file1_df.shape)
print('File2 shape:', file2_df.shape)
print('File3 shape:', file3_df.shape)
```

Standardize header names

```python
# We check if header names are different
print('File1 w/ file2', file1_df.columns == file2_df.columns)
print('File2 w/ file3', file2_df.columns == file3_df.columns)
print('File1 w/ file3', file1_df.columns == file3_df.columns,'\n')

# Let's see what's different...
print('File1:', list(file1_df.columns))
print('File2:', list(file2_df.columns))
print('File3:', list(file3_df.columns),'\n')

# Harmonize header names
file1_df = file1_df.rename(columns={'ST':'State','GENDER':'Gender'})
file2_df = file2_df.rename(columns={'ST':'State','GENDER':'Gender'})

# Lower header names
file1_df.columns = map(lambda header : header.lower(), file1_df.columns)
file2_df.columns = map(lambda header : header.lower(), file2_df.columns)
file3_df.columns = map(lambda header : header.lower(), file3_df.columns)

# Replace spaces for _
file1_df.columns = map(lambda x : x.replace(" ", "_"), file1_df.columns)
file2_df.columns = map(lambda x : x.replace(" ", "_"), file2_df.columns)
file3_df.columns = map(lambda x : x.replace(" ", "_"), file3_df.columns)


print('File1 standardized:', list(file1_df.columns))
print('File2 standardized:', list(file2_df.columns))
print('File3 standardized:', list(file3_df.columns))
```

Rearrange the columns in the dataframe as needed

```python
file1_df = file1_df[['customer', 'state', 'gender', 'education', 'customer_lifetime_value', 
                     'income', 'monthly_premium_auto', 'number_of_open_complaints', 
                     'total_claim_amount', 'policy_type', 'vehicle_class']]
file2_df = file2_df[['customer', 'state', 'gender', 'education', 'customer_lifetime_value', 
                     'income', 'monthly_premium_auto', 'number_of_open_complaints', 
                     'total_claim_amount', 'policy_type', 'vehicle_class']]
file3_df = file3_df[['customer', 'state', 'gender', 'education', 'customer_lifetime_value', 
                     'income', 'monthly_premium_auto', 'number_of_open_complaints', 
                     'total_claim_amount', 'policy_type', 'vehicle_class']]

#Check if all headers are equal
print(file1_df.columns == file2_df.columns)
print(file2_df.columns == file3_df.columns)
```

Concatenate the three dataframes

```python
data = pd.concat([file1_df, file2_df, file3_df], axis=0).reset_index(drop=True)
display(data)
```

Which columns are numerical?

```python
print(data.dtypes, '\n')
print('Numerical columns are:', list(data._get_numeric_data().columns))
```

Which columns are categorical?

```python
print('Categorical columns are:', list(data.select_dtypes(['object']).columns))
```

Delete the column education and the number of open complaints from the dataframe.

```python
data = data.drop(['education', 'number_of_open_complaints'], axis = 1)
display(data)
```

Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

```python
#Droping nan values
print('Number of nan values before', data['customer_lifetime_value'].isna().sum())
data = data[data['customer_lifetime_value'].isna()==False]
print('Number of nan values after:', data['customer_lifetime_value'].isna().sum())

def clean_data (df_cols):
    data_transformed = []
    for index, value in enumerate(df_cols):
        if isinstance (value, str):
            value = value.replace('%', "")
            data_transformed.append(float(value))
        else:
            data_transformed.append(value)
    return data_transformed
        
data['customer_lifetime_value'] = data[['customer_lifetime_value']].apply(clean_data)
data['customer_lifetime_value'].head(20)
```

Check for duplicate rows in the data and remove if any.

```python
print('Data before droping duplicates:', data.shape)
data = data.drop_duplicates()
print('Data after droping duplicates:', data.shape)
```

Filter out the data for customers who have an income of 0 or less.

```python
low_income = data[data['income'] <= 0].reset_index(drop=True)
display(low_income)
```
